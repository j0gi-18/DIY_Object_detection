```python
# code to reorganize dataset
import os
import shutil

def organize_files(source_directory):
    # Define target directories
    annotations_folder = os.path.join(source_directory, 'Annotations')
    jpeg_images_folder = os.path.join(source_directory, 'JPEGImages')

    # Create target directories if they don't exist
    os.makedirs(annotations_folder, exist_ok=True)
    os.makedirs(jpeg_images_folder, exist_ok=True)

    # Loop through files in the source directory
    for filename in os.listdir(source_directory):
        source_path = os.path.join(source_directory, filename)

        # Check if it is a file
        if os.path.isfile(source_path):
            file_extension = filename.lower().split('.')[-1]  # Get the file extension

            # Move files based on their extensions
            if file_extension == 'xml':
                shutil.move(source_path, annotations_folder)
                print(f'Moved: {filename} to {annotations_folder}')
            elif file_extension in ['jpg', 'jpeg', 'png']:
                shutil.move(source_path, jpeg_images_folder)
                print(f'Moved: {filename} to {jpeg_images_folder}')

# Example usage: specify the path to the folder containing mixed files
source_path = '/content/keras-retinanet/dataset/train/train_again'
organize_files(source_path)
```

```python
import os
import shutil

def move_all_files(source_directory, destination_directory):
    # Ensure the destination directory exists
    os.makedirs(destination_directory, exist_ok=True)

    # List all files in the source directory
    for filename in os.listdir(source_directory):
        source_path = os.path.join(source_directory, filename)

        # Check if it's a file (not a directory)
        if os.path.isfile(source_path):
            # Move the file to the destination directory
            shutil.move(source_path, destination_directory)
            print(f'Moved: {filename} to {destination_directory}')

# Example usage: specify the path to the source and destination folders
source_path = '/content/keras-retinanet/dataset/train/train_again/JPEGImages'
destination_path = '/content/keras-retinanet/dataset/JPEGImages'
move_all_files(source_path, destination_path)
```

```python
from os import listdir
from os.path import isfile, join

dataset_directory = "./dataset/Annotations"

files = [join(dataset_directory,file) for file in listdir(dataset_directory) if isfile(join(dataset_directory,file))]

i = 0
with open("test.txt", "w") as test:
    for file in files:
        i += 1
        if (i % 5) == 0:
            name = file.split("\\")[-1]
            name = name.split(".")[0]
            test.write(name + "\n")
            print("test:    " + name)

i = 0
with open("train.txt", "w") as train:
    for file in files:
        i += 1
        if (i % 5) != 0:
            name = file.split("\\")[-1]
            name = name.split(".")[0]
            train.write(name + "\n")
            print("train:    " + name)

test.close()
train.close()
```

```python
import os

def count_files_in_folder(folder_path):
    try:
        # List all files in the specified folder
        files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

        # Return the count of files
        return len(files)
    except FileNotFoundError:
        print(f"The folder '{folder_path}' does not exist.")
        return 0
    except Exception as e:
        print(f"An error occurred: {e}")
        return 0

# Example usage: specify the path to the folder
folder_path = input("Enter the path to the folder: ")
total_files = count_files_in_folder(folder_path)

print(f"Total number of files in '{folder_path}': {total_files}")
```
